```python
import os
import csv
import torch
from ultralytics import YOLO

# ==============================
# CONFIG
# ==============================
model_path = "/kaggle/working/runs/classify/train/weights/best.pt"   # your trained model
image_dir = dataset2.location + "/train/0 - test"           # directory of test images
csv_file = "test_predictions_yolo.csv"

# ==============================
# LOAD MODEL
# ==============================
device = "cuda" if torch.cuda.is_available() else "cpu"
model = YOLO(model_path)

# ==============================
# PREDICTION LOOP
# ==============================
results = []

with torch.no_grad():
    predictions = model.predict(image_dir, verbose=False)

    for pred in predictions:
        filename = os.path.basename(pred.path)

        # ✅ Clean up Roboflow suffix like ".rf.xxxxx"
        if ".rf." in filename:
            filename = filename.split(".rf.")[0]

        # ✅ Replace "_jpeg" suffix with ".jpeg"
        if filename.endswith("_jpeg"):
            filename = filename.replace("_jpeg", ".jpeg")

        # Top-1 predicted class (ID instead of name)
        top1_id = int(pred.probs.top1)

        results.append([filename, top1_id])

# ==============================
# SAVE CSV (no index)
# ==============================
with open(csv_file, mode="w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Id", "Predictions"])  # header
    writer.writerows(results)

print(f"✅ Predictions saved to {csv_file}")
```
